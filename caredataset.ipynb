# **Data Preprocessing:**

```python
import pandas as pd
# import the data from csv file to correct changes
# data_befor=pd.read_csv("B:\PUA\year 1_2 - semstar 2\DS - DS 102\project\caredataset.csv")
data_befor=pd.read_csv("/content/caredataset.csv")
# Handle missing values
print(data_befor.isnull().sum())
#there is no missing values
```

```python
# Detect and treat outliers
# Detect
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(3,3))
plt.title('befor treat outliers')
print(sns.boxplot(data_befor["Age"])) #we can see that whata above 65 is outliers
# treat outliers
data = data_befor[data_befor["Age"] <= 65]
plt.figure(figsize=(3,3))
plt.title("after treat outliers")
print(sns.boxplot(data["Age"]))
```

```python
# remove un uesed columns
# index & Patient Id
data.drop(columns=['index','Patient Id'], inplace=True)
print(data)
```

```python
# Encode categorical variables into numerical format
# we will encode level column on scale Low: 0,Medium: 1 ,High:3
data["Level"] = data["Level"].map({"Low": 0, "Medium": 1 ,"High":2})
print(data["Level"])
```

```python
# feature engineering
# Selection & Extraction of features
# we will remove OccuPational Hazards  , Clubbing of Finger Nails columns
data.drop(columns=['OccuPational Hazards','Clubbing of Finger Nails'], inplace=True)


```

```python
# Scale numerical features to a similar range
from sklearn import preprocessing as pp
import numpy as np
data_scaled=np.array([(data["Chest Pain"]),(data["Fatigue"]),data["Shortness of Breath"]])
data_scaled=pp.MinMaxScaler(feature_range=(1,7)).fit_transform(data_scaled)
data["Chest Pain"]=data_scaled[0]
data["Fatigue"]=data_scaled[1]
data["Shortness of Breath"]=data_scaled[2]
for col in data.columns:
    print(col,"\tmax=",max(data[col]),"\tmin=",min(data[col]))
```

```python
# Split dataset into training and testing sets
# training_data
# print(data.shape) #990 row ---> 793 row training
training_data=data.iloc[ :793,:]
print(training_data.describe())
print("#"*100)
print(training_data.info())
```

```python
# testing_data
# print(data.shape) #990 row ---> 198 row  testing
testing_data=data.iloc[793: ,:]
print(testing_data.describe())
print("#"*100)
print(testing_data.info())
```

# **Data modeling**

```python
# Train & Test Splitting the Data
from sklearn.model_selection import train_test_split
```

```python
x_data=data.drop(columns=["Level"])
y_data=data["Level"]
x_data_train,x_data_test,y_data_train,y_data_test=train_test_split(x_data,y_data,test_size=0.2,random_state=42)
```

```python
from sklearn.metrics import precision_score, recall_score, accuracy_score, r2_score,f1_score, confusion_matrix, ConfusionMatrixDisplay, classification_report
```

```python
def train_Metrics(x_data_train_pre):
    print("Train Data Metrics:")
    print("Precision : ", precision_score(y_data_train, x_data_train_pre, average='micro'))
    print("Recall : ", recall_score(y_data_train, x_data_train_pre, average='micro'))
    print("Accuracy : ", accuracy_score(y_data_train, x_data_train_pre,))
    print("F1 Score : ", f1_score(y_data_train, x_data_train_pre, average='micro'))
    print("R2 Score : ", r2_score(y_data_train, x_data_train_pre))

```

```python
def test_Metrics(x_data_test_pre):
    print("Train Data Metrics:")
    print("Precision : ", precision_score(y_data_test, x_data_test_pre, average='micro'))
    print("Recall : ", recall_score(y_data_test, x_data_test_pre, average='micro'))
    print("Accuracy : ", accuracy_score(y_data_test, x_data_test_pre,))
    print("F1 Score : ", f1_score(y_data_test, x_data_test_pre, average='micro'))
    print("R2 Score : ", r2_score(y_data_test, x_data_test_pre))
    print("classification_report",classification_report(y_data_test, x_data_test_pre))
    print("confusion_matrix", confusion_matrix(y_data_test, x_data_test_pre))
    ConfusionMatrixDisplay(confusion_matrix =confusion_matrix(y_data_test, x_data_test_pre), display_labels=['Low', 'Medium', 'High']).plot()
```

```python
from sklearn.linear_model import LogisticRegression
model_lr = LogisticRegression()
model_lr.fit(x_data_train, y_data_train)
```

```python
x_data_train_pre=model_lr.predict(x_data_train)
train_Metrics(x_data_train_pre)
```

```python
x_data_test_pre=model_lr.predict(x_data_test)
test_Metrics(x_data_test_pre)
```

```python
print('Logistic Regression Score = ',model_lr.score(x_data_test,y_data_test))
```

```python
from sklearn.ensemble import AdaBoostClassifier
model_ADABboost=AdaBoostClassifier()
model_ADABboost.fit(x_data_train,y_data_train)
```

```python
x_data_train_pre=model_ADABboost.predict(x_data_train)
train_Metrics(x_data_train_pre)
```

```python
x_data_test_pre=model_ADABboost.predict(x_data_test)
test_Metrics(x_data_test_pre)
```

```python
print('AdaBoostClassifier Score = ',model_ADABboost.score(x_data_test,y_data_test))
```
